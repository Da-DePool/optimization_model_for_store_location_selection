```python
df = pd.read_excel('ventas_oxxo.xlsx', sheet_name='pedidos_davis')
df
```

```python
def calcular_tienda_venta(df, tienda):
    df_tienda = df[df['nombre_tienda'] == tienda].copy()
    total_venta_tienda = df_tienda['monto'].sum()
    return total_venta_tienda
```
